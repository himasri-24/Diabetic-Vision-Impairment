```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Simulated accuracy and loss values based on the given training logs

# CNN Model (Custom Model)
cnn_train_acc = [0.4649, 0.6057, 0.6277, 0.6632, 0.6789]
cnn_val_acc = [0.6364, 0.6745, 0.6782, 0.6455, 0.6800]
cnn_train_loss = [1.3908, 1.0462, 0.9864, 0.9209, 0.8971]
cnn_val_loss = [1.0053, 0.9247, 0.9070, 0.9249, 0.8507]

# VGG16 Model
vgg_train_acc = [0.4946, 0.5907, 0.6097, 0.6216, 0.6245]
vgg_val_acc = [0.6255, 0.6345, 0.6236, 0.6473, 0.6255]
vgg_train_loss = [1.4851, 1.0990, 1.0604, 1.0272, 1.0092]
vgg_val_loss = [1.0134, 0.9405, 0.9756, 0.9227, 0.9157]

epochs = range(1, 6)

# Plot CNN Accuracy and Loss Graphs
plt.figure(figsize=(12, 5))

# CNN Accuracy
plt.subplot(1, 2, 1)
plt.plot(epochs, cnn_train_acc, 'bo-', label='Train Accuracy')
plt.plot(epochs, cnn_val_acc, 'b--', label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('CNN Model Accuracy')
plt.legend()

# CNN Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, cnn_train_loss, 'bo-', label='Train Loss')
plt.plot(epochs, cnn_val_loss, 'b--', label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('CNN Model Loss')
plt.legend()

plt.show()

# Plot VGG16 Accuracy and Loss Graphs
plt.figure(figsize=(12, 5))

# VGG16 Accuracy
plt.subplot(1, 2, 1)
plt.plot(epochs, vgg_train_acc, 'ro-', label='Train Accuracy')
plt.plot(epochs, vgg_val_acc, 'r--', label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('VGG16 Model Accuracy')
plt.legend()

# VGG16 Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, vgg_train_loss, 'ro-', label='Train Loss')
plt.plot(epochs, vgg_val_loss, 'r--', label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('VGG16 Model Loss')
plt.legend()

plt.show()

# Simulated Confusion Matrices
y_true = np.random.randint(0, 5, 100)  # Simulated true labels (5 classes)
y_pred_cnn = np.random.randint(0, 5, 100)  # Simulated CNN predictions
y_pred_vgg = np.random.randint(0, 5, 100)  # Simulated VGG16 predictions

# Compute confusion matrices
cm_cnn = confusion_matrix(y_true, y_pred_cnn)
cm_vgg = confusion_matrix(y_true, y_pred_vgg)

# Plot confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(cm_cnn, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('CNN Confusion Matrix')
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')

sns.heatmap(cm_vgg, annot=True, fmt='d', cmap='Reds', ax=axes[1])
axes[1].set_title('VGG16 Confusion Matrix')
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')

plt.show()

# Classification Reports
report_cnn = classification_report(y_true, y_pred_cnn, output_dict=True)
report_vgg = classification_report(y_true, y_pred_vgg, output_dict=True)

# Extract precision, recall, and F1-score for both models
cnn_precision = np.mean([report_cnn[str(i)]['precision'] for i in range(5)])
cnn_recall = np.mean([report_cnn[str(i)]['recall'] for i in range(5)])
cnn_f1 = np.mean([report_cnn[str(i)]['f1-score'] for i in range(5)])

vgg_precision = np.mean([report_vgg[str(i)]['precision'] for i in range(5)])
vgg_recall = np.mean([report_vgg[str(i)]['recall'] for i in range(5)])
vgg_f1 = np.mean([report_vgg[str(i)]['f1-score'] for i in range(5)])

# Final Comparison
cnn_final_acc = cnn_val_acc[-1]
vgg_final_acc = vgg_val_acc[-1]

cnn_metrics = {"Accuracy": cnn_final_acc, "Precision": cnn_precision, "Recall": cnn_recall, "F1-Score": cnn_f1}
vgg_metrics = {"Accuracy": vgg_final_acc, "Precision": vgg_precision, "Recall": vgg_recall, "F1-Score": vgg_f1}

print("CNN Model Metrics:", cnn_metrics)
print("VGG16 Model Metrics:", vgg_metrics)
```
